```python
import pandas as pd
```

```python
random_bubble_data = pd.read_csv('float_array_bubble_sort.csv')
random_bubble_data['float_list_bubble_sort'] = pd.read_csv('float_list_bubble_sort.csv')
random_bubble_data['int_array_bubble_sort'] = pd.read_csv('int_array_bubble_sort.csv')
random_bubble_data['int_list_bubble_sort'] = pd.read_csv('int_list_bubble_sort.csv')
random_bubble_data['dimension'] = pd.Series(range(1, 1000, 20))
random_bubble_data
```

```python
random_bubble_data.plot(title='Bubble sort for random array',x='dimension', grid=True)
```

```python
random_bubble_data.loc[:,['int_array_bubble_sort', 'float_array_bubble_sort', 'dimension']].plot(
    x='dimension', title='Int and Float array bubble sort', grid=True)
```

```python
random_quick_data = pd.read_csv('float_array_quick_sort.csv')
random_quick_data['float_list_quick_sort'] = pd.read_csv('float_list_quick_sort.csv')
random_quick_data['int_array_quick_sort'] = pd.read_csv('int_array_quick_sort.csv')
random_quick_data['int_list_quick_sort'] = pd.read_csv('int_list_quick_sort.csv')
random_quick_data['dimension'] = pd.Series(range(1, 1000, 20))
random_quick_data
```

```python
random_quick_data.plot(x='dimension', title='Quick random sorts', grid=True)
```

```python
random_quick_data.loc[:,['int_array_quick_sort', 'float_array_quick_sort', 'dimension']].plot(
    x='dimension', title='Int and Float array quick sort', grid=True)
```

Сравнение быстрой и пузырьковой сортировок

```python
bubble_quick_random_comparing = random_bubble_data.merge(random_quick_data, on='dimension', how='outer')
bubble_quick_random_comparing
```

Сравнение пузырьковой и быстрой сортировок на целочисленном массиве

```python
bubble_quick_random_comparing.loc[:,['int_array_bubble_sort', 'int_array_quick_sort', 'dimension']].plot(
    x='dimension', title='Быстрая и пузырьковая сортировки на случайном массиве', grid=True)
```

```python
bubble_quick_random_comparing.loc[:,['int_list_bubble_sort', 'int_list_quick_sort', 'dimension']].plot(
    x='dimension', title='Быстрая и пузырьковая сортировки на случайном списке', grid=True
)
```
